Via Hebbian learning hand neurons in the first SOM are connected to joint neurons in the second SOM.The "real" quality of Hebbian learning can be tested in human-robot interaction experiments, where Nao has to follow an object. In that setting, Nao sees marker coordinates and tries to find the closest coordinates in his SOM storing hand weights.Based on the activated neuron in the first SOM, a neuron in the second SOM gets activated and Nao sets his joint coordinates correspondingly.

## Training networksReading of knowledge bases, training and saving configuration files is done in the [minisom_nao.py](https://github.com/ikajic/pointingNao/blob/master/python/minisom_nao.py) script.By default, networks are trained on the longest babbling data set, the one that took 42 minutes. I trained many different network configurations which differ in:1. Number of neurons: either 5x5, 10x10, 15x152. Number of epochs used for training: almost always 20.000, but in certain cases less (200 epochs) to show examples of "undertrained" network3. Number of data points used in training:  * In some simulations I used only first `n` points which simulate very short learning/babbling process * In others I used every `n`-th point, to simulate the babbling process which was slightly "sped-up" * There are some simulations where I used all available samples in knowledge baseWhich configuration was used for a particular instance of a trained network can be seen in the directory name that contains configuration files for that network.All configurations are stored in the `somconfig\` directory and folders there have a name format **nxn_e_mode**.Within each such folder, additional information on training can be found in `out.log` and `param.json` files.* **nxn** is the size of a SOM. For example if a network as 5 columns and 5 rows, the overall number of neurons in that SOM is 25* **e** is the number of epochs used to train a network* **mode** string contains two parts (`lp`), a letter `l` which can be `f` or `u` and a number `p`. The letter defines how training data were chosen from a training set. `f` means only first n points were used, `u` means the data were uniformly chosen such that each n-th point was choosen. n is computed from the number `p` which denotes percentage.For example, if the directory name is: `5x5_20000_f1` that means it contains configuration for a SOM with 25 neurons that was trained on 20.000 epochs and only first 1% of data from a training set were used for the training.Let's take a look at the SOM beauty pageant.
