# Testing different SOM configurations 

## Data: Motor babbling using RW
I ran multiple babbling sessions with Nao using only the left hand and for each session I saved the knowledge base.
In the knowledge base each row contains 3D hand coordintes and 4D joint coordinates (2 for elbow and 2 for shoulder). 
All knowledge bases I collected so far are stored in the [data](https://github.com/ikajic/pointingNao/tree/master/data) directory.
For purposes of demonstration, I used the two longest babbling sessions.
Here are some examples of how hand coordinates obtained during these babbling phases look like. 
The first plot shows data from a babbling phase that took place 37 minutes and yielded ~74.000 samples. 
This data are also used to train networks afterwards.

```python
import numpy as np
import matplotlib.pyplot as pl
from dbg import plot3, fetch_som

# This data set will be used for training of networks covering the 'hand space' later
_37min_hands = np.genfromtxt('../data/r_37min_74k.dat', skip_header=2, usecols=(0,1,2))
print "Number of samples:", _37min_hands.shape
plot3(_37min_hands)
show()
```

This data points are also generated in webots and babbling session lasted for 29 minutes. This set is used to test accuracy of networks later on.

```python
_29min = np.genfromtxt('../data/r_29min_58k.dat', skip_header=2, usecols=(0,1,2))
print "Number of samples:", _29min.shape
plot3(_29min)
show()
```

And these are joint coordinates (with 2 degrees of freedom for both the elbow and the shoulder) and they don't look particularly intuitive when plotted. 
Nevertheless, they are used to train the second SOM network, which has 4D neurons. 

```python
_37min_joints = np.genfromtxt('../data/r_37min_74k.dat', skip_header=2, usecols=(3,4,5,6))
figure()
subplot(121)
scatter(_37min_joints[:,0], _37min_joints[:,1])
title('Elbow')
subplot(122)
scatter(_37min_joints[:,2], _37min_joints[:,3])
title('Shoulders')
show()
```

Via Hebbian learning hand neurons in the first SOM are connected to joint neurons in the second SOM.
The "real" quality of Hebbian learning can be tested in human-robot interaction experiments, where Nao has to follow an object. 
In that setting, Nao sees marker coordinates and tries to find the closest coordinates in his SOM storing hand weights.
Based on the activated neuron in the first SOM, a neuron in the second SOM gets activated and Nao sets his joint coordinates correspondingly.

## Training networks
Reading of knowledge bases, training and saving configuration files is done in the [minisom_nao.py](https://github.com/ikajic/pointingNao/blob/master/python/minisom_nao.py) script.
By default, networks are trained on the longest babbling data set, the one that took 42 minutes. 
I trained many different network configurations which differ in:

1. Number of neurons: either **5x5**, **10x10**, **15x15**
2. Number of epochs used for training: almost always 20.000, but in certain cases less (200 epochs) to show examples of "undertrained" network
3. Number of data points used in training: 
 * In some simulations I used only first `n` points which simulate very short learning/babbling process
 * In others I used every `n`-th point, to simulate the babbling process which was slightly "sped-up"
 * There are some simulations where I used all available samples in knowledge base

## Understanding directory names
Which configuration was used for a particular instance of a trained network can be seen in the directory name that contains configuration files for that network.
All configurations are stored in the `somconfig\` directory and folders there have a name format **nxn_e_mode**.
Within each such folder, additional information on training can be found in `out.log` and `param.json` files.

* **nxn** is the size of a SOM. For example if a network as 5 columns and 5 rows, the overall number of neurons in that SOM is 25
* **e** is the number of epochs used to train a network
* **mode** string contains two parts (`lp`), a letter `l` which can be `f` or `u` and a number `p`. 
The letter defines how training data were chosen from a training set. 
`f` means only first n points were used, `u` means the data were uniformly chosen such that each n-th point was choosen. n is computed from the number `p` which denotes percentage.

For example, if the directory name is: `5x5_20000_f1` that means it contains configuration for a SOM with 25 neurons that was trained on 20.000 epochs and only first 1% of data from a training set were used for the training.

## Testing networks
To test networks and compare them among each other I used data from the babbling session that lasted 34 minutes and is plotted above. 
First, fetch the data:

```python
from minisom_nao import read_data

test_path = "../data/r_29min_58k.dat"
    
# 100 as argument means use 100% of data in *.dat folder
test_data = read_data(test_path, 100, 'u')['hands']
print "Testing using ", test_data.shape, " points"
```

Now, we will present a network with all the 3D vectors in `test_data` and for each entry find the neuron that approximates this vector. 
Each vector corresponds to 3D hand coordinates (x, y, z). 
We will compute the average error and the standard deviation for each network.
We use following networks:

```python
somconfs = ["5x5_20000_f1",
            "5x5_20000_f10",
            "5x5_20000_u10",
            "5x5_20000_u100",
            "10x10_20000_f1",
            "10x10_20000_f10",
            "10x10_20000_u10",
            "10x10_20000_u100",
            "15x15_20000_f1",
            "15x15_20000_f10",
            "15x15_20000_u10",
            "15x15_20000_u100"]
```

Now, we call `prediction_error` function from the [test_som.py](https://github.com/ikajic/pointingNao/blob/master/python/test_som.py) script which prettyprints errors and standard deviations. 
In `errors`, for each network we store the prediction error for each individual data point (hand vector) in test_data.

```python
from test_som import prediction_errors

errors = prediction_errors(test_data, somconfs)    
```

### T-tests
We run dependent paired t-tests on error data for each possible combination of two networks. 
[T-test](http://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.mstats.ttest_rel.html) function is available in the scipy.stats library.
The implementation of the function below, `table_print_errors` can be found in [test_som.py](https://github.com/ikajic/pointingNao/blob/master/python/test_som.py).

```python
from scipy.stats import ttest_rel
from test_som import table_print_errors

table_print_errors(errors, somconfs)
```

### Testing SOMs with data sets of different sizes

```python
from test_som import generate_test_set

nr_pts = np.array([1, 10, 25])*1000                              
nr_trials = 1

soms = []
for path in somconfs:
    soms.append(fetch_som('somconf/'+path+'/'))
    
errors = np.zeros((len(soms), len(nr_pts), nr_trials))

for i in xrange(nr_trials):
    for j, pts in enumerate(nr_pts):
        data = generate_test_set(test_data, pts)
        print pts, " points"
        for k in np.arange(len(somconfs)):
            err = soms[k].quantization_error(data)
            print "SOM ", somconfs[k], " error: ", err.mean()
            errors[k,j,i] = err.mean()
            
```
